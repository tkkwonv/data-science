포아송 분포(Poisson distribution)는 사람마다 '포와송 분포, 푸아송 분포'라고 부르기도 합니다.  
포아송 분포는 제 생각에 아마도 이산 확률분포에서 가장 많이 사용하는 분포가 아닐까 싶습니다.

포아송 분포를 알아보기 전에 '모수(population parameter)'를 먼저 이해하셔야 합니다.  
확률 및 통계학에서 모수는 '모집단의 특성을 나타내는 수치' 입니다. 뭔가 좀 애매한 느낌이 들죠?  
포아송 분포에 한정해서 모수를 설명하자면, 한 시간 동안 은행에 다녀간 고객의 수, 한 시간 동안 사무실에 걸려온 전화의 수, 또는 어떤 책의 한 페이지에 존재하는 오타의 수 등을 나타내는 숫자를 얘기 합니다.  
즉, 포아송 분포에서의 모수는 '단위시간 또는 단위공간에서 평균발생횟수'를 의미합니다.  
포아송 분포에서 모수는 람다(λ)로 표시 합니다. 따라서 '포아송 분포'는 단위 시간, 단위 공간에 어떤 사건이 몇 번 발생할 것인가를 표현하는 이산 확률분포입니다.

포아송 분포에서의 확률은 모수에 따라 그 결과가 달라집니다.  
어떤 사건의 발생횟수가 (모수를 갖는) 포아송 분포를 따르는 확률과정을 '포아송 과정 또는 포아송 프로세스(Poisson Process)'라 하고, 여기서 '확률과정(또는 랜덤과정, random process)'이라는 말은 시간과 관련된 확률적인 성격을 갖는 것을 의미합니다.  
즉, 간단히 말하면 시간에 따라 확률도 변한다고 이해하시면 됩니다. 따라서 '포아송 과정'도 시간에 따라 그 확률이 변한다고 이해하시면 되겠죠.

포아송 분포의 전제조건은 다음과 같습니다.

1. 독립성 : 어떤 단위 시간 또는 단위 공간에서 발생한 결과는 중복되지 않은 다른 시간이나 공간에서 발생한 결과와 서로 독립이다.

2. 일정성 : 어떤 단위 시간 또는 단위 공간에서 발생한 확률(또는 횟수)은 그 시간의 크기, 혹은 공간의 크기에 비례하고, 외부의 영향을 받지 않는다. 즉 단위 시간이나 공간에서 발생한 평균발생횟수는 일정하다.

3. 비집락성 : 매우 짧은 시간이나 매우 작은 공간에 두 개 이상의 결과가 동시에 발생할 확률은 0 이다.

위 조건을 간단히 예를 들어 설명하면,

1. 9시부터 10시까지 하나은행에 들른 고객의 수와 같은 시간 동안 국민은행에 들른 고객의 수는 서로 독립이고, 또한 9시부터 10시까지 하나은행에 들른 고객의 수와 10시부터 11시까지 신한은행에 들른 고객의 수도 역시 서로 독립이라는 의미입니다.

2. 만약 어떤 사건이 1분에 평균 2번 발생한다면, 3분 동안에는 평균 6번이 발생한다는 의미입니다.

3. 서해안고속도로에 같은 시간에 같은 지점에서 같은 교통사고가 두 번 이상 발생할 확률은 무시해도 좋다는 의미입니다.


참고로 확률 및 통계학에서 용어에 '랜덤(random)'이란 단어가 굉장히 많습니다. 확률통계학에서 말하는 '랜덤'의 의미는 '시간적으로 미리 결과에 대해 정확히 예측할 수 없다'는 것을 의미합니다. 그리고 '과정(process)'의 의미는 '시간과 관련된 상태'를 말할 때 주로 사용합니다.  
시간과 관련되지 않은 상태의 확률을 구할 때는 '사건(event)'이라는 단어를 씁니다.

위에서 포아송 분포는 단위 시간 혹은 단위 공간 안에 어떤 사건이 몇 번 발생할 것인가를 표현하는 확률 분포라고 했습니다.  
이때, 포아송 분포의 확률변수 X를 '포아송확률변수(poisson random variable)'라고 합니다. 

확률질량함수는 이산확률변수의 확률함수입니다


보통 다음과 같은 확률 문제를 알아내기 위해 포아송 분포를 많이 응용하여 사용합니다.

* 하루동안 발생하는 출생자 수 혹은 사망자 수
* 어떤 특정 진도 이상의 지진이 발생하는 수
* 어떤 특정량의 방사선을 DNA에 쬐었을 때 발생하는 돌연변이 수
* 어떤 시간 동안 톨게이트를 통과하는 차량의 수
