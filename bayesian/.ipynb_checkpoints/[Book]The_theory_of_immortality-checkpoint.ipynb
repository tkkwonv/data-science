![불멸의이론](http://ncc.phinf.naver.net/20160811_127/14708926306679uV7k_PNG/v48_06_1.png)

                 모든 지식은, 끝까지 분석해 들어가면, 역사이다.
                 모든 과학은, 추상적으로 말하면, 수학이다.
                 모든 판단은, 원리를 따져보면, 통계학이다.
                 All knowledge is, in the final analysis, history.
                 All sciences are, in the abstract, mathematics.
                 All judgments are, in their rationale, statistics.

                                                                        - 조재근 역, <통계학의 역사> 서문에서

위 문구는 개인적으로 자주 인용하는 말이다. 지식은 역사에 관한 것이지만 삶에 필요한 모든 판단은 통계학적이다라는 말은 오늘 소개하고자 하는 책, <불멸의 이론>이 말하고자 하는 세계관과 맞닿아 있다.

<불멸의 이론>에 관한 서평은 일년을 두고 미루어 왔다. 거기에는 몇 가지 이유가 있었다. 무엇보다 이 책에서 다루는 ‘베이지안(Bayesian) 방법론’을 수학적으로 쉽게 설명할 자신이 없었다. 둘째, 학술 영역에서 통용될 내용이라 일반인들에게 소개하는 것이 마땅치 않을 것 같았다. 또 꽤 오래 전(2013년)에 소개된 책이라 시의성도 그리 크지 않다고 생각했다.

하지만 매번 서평을 쓰려 할 때마다 책꽂이 한 켠에 자리잡고 있는 이 책이 눈에 밟혔다. 왜 그럴까? 돌이켜 생각해 보면 이 책을 읽게 된 계기와 이 책에서 얻었던 개인적 소회를 그 누군가와 꼭 나누고 싶었던 ‘허영심’이 책에 대한 이해부족에서 오는 저어하는 마음보다 컸던 것이 아닐까 싶다.

그래서 이 책에 대한 서평은 책의 내용보다는 개인적인 소회를 중심으로 기술하고자 한다. 먼저 책과의 만남이다.

##### 왜 베이지안(Bayesian) 인가?

한 동안 이 코너를 통해 ICT에 관련된 책을 다수 소개했다. 그 중에 한 권이 네이트 실버(Nate Silver)의 <신호와 소음>이다. <신호와 소음>의 서평1) 을 읽어 보신 분은 눈치채셨겠지만 이 책에서 중요한 부분을 차지하는 8장 ‘베이즈 정리’에 대한 자세한 소개는 책을 읽어 보는 것이 좋다는 톤으로 정리하고 말았다. 이 8장을 읽다 보면 베이즈 정리를 설명하는 내용에 달린 미주에 <불멸의 이론>이 소개 되어 있음을 발견할 것이다.

하지만 미주 속에 인용된 그 많은 책 중에서 왜 이 책에 특히 더 관심이 갔을까? 그 이유는 알고리즘에 대한 관심 때문이었다. 2007년 애플이 아이폰을 소개한 이후로, ‘Next Big Thing’으로서 ICT 업계를 사로잡은 화두는 빅데이터(Big data)였다. 다음에는 IoT였고 그 다음이 알고리즘, 기계학습, 인공지능으로 이어졌다.

이 부분을 좀 더 풀어서 써보면 다음과 같다. 빅데이터는 새로운 경영자원으로 각광을 받았고 구글의 많은 서비스들, 자동번역, 검색어 예측 등이 빅데이터를 통해서 제공되고 있다는 점이 소개 되면서 전세계적인 관심을 받았다. 이후 보다 많은 데이터를 위한 Datafication, 즉 모든 것들에 센서를 심어 데이터를 양산하자는 아이디어가 IoT로 발전해 나갔다. 하지만 여기에서 초점은 센서나 단말 같은 하드웨어가 아니고 거기에서 생산된 데이터에서 정보를 추출하고 이를 서비스로 연결시키는 소프트웨어로 모아졌다. 이 부분을 좀 더 기계적으로 처리하는 방식으로 알고리즘이 부상했고 데이터의 투입에 따라 알고리즘이 ‘진화’한다는 부분에서 바로 ‘베이지안’ 통계 기법이 핵심적 방법론으로 부각된다. (기계학습과 인공지능은 알고리즘 진화의 결과물들이다.) 따라서 iPhone과 같은 스마트폰이 우리 생활 속으로 깊숙이 스며들어 온 후, 우리가 향유하고 있는 수 많은 서비스들을 통해, 베이지안 통계는 우리의 삶을 지배하고 있다고 할 수 있다. 따라서 ‘그렇다면 베이지안 통계는 무엇인가’라는 질문에 도달한 것은 ICT 업계를 지켜보는 사람으로서 자연스러운 귀결점이었다.

##### 베이지안 통계란?

베이즈는 1701년 영국 남서부 하트퍼드셔 카운티에서 태어났다. 에든버러 대학을 졸업하고 왕립학회 회원으로도 선출되었다. 그가 베이즈 정리를 고안하게 된 계기는 회의론자인 데이비드 흄(David Hume)이 인과관계에 대해 의문을 제시한데 있었다. 흄은 모든 것의 제1원인이 신이라는 믿음에 반하게도 ‘세상이 현재의 모습처럼 설계되어 있다는 사실’이 궁극 원인인 창조자의 존재를 입증하는 것은 아니라고 주장했다. 이는 기독교의 핵심적인 믿음을 훼손하는 것이었다.

흄이 원인과 결과에 대해 의심을 품고 나서자 베이즈는 이 문제를 수학적으로 처리할 수 있는 방법을 모색하기 시작했다. 그리고 다음과 같은 목표를 세운다. 즉 <과거에 대해서만, 즉 어떤 일(사건)이 일어난 횟수 혹은 일어나지 않은 횟수에 대해서만 알고 있을 때, 아무 것도 모르는 미래에 일어날 사건의 확률 근사치를 알아내는 것>이 바로 그것이다. 이 문제를 해결하기 위해 그는 시작점으로 어떤 숫자를 설정한 후, 보다 많은 정보를 수집하며 그 숫자를 보정하는 방법을 취한다. 이를 개념적으로 제시한 것이 당구대 실험이다. (<신호와 소음> 서평 참조).

![당구대](http://ncc.phinf.naver.net/20160812_227/14709833362802Ie0q_PNG/v48_06.png)

    정사각형 테이블 ABCD와 공W와 O가 있는 이른바 베이스의 “당구대”
    공 W를 굴려 이것이 선 os 위에 멈추게 되면 이어서 공 O를 n차례 굴려 그 공이 사각형 osDA 안에서 멈추는 횟수를 기록한다. 아래편의 곡선은 횟수를 기록     한 후 얻게 되는 비율 Ao/AB의 사후밀도함수를 나타낸다. 이 밀도함수를 베이스가 위아래로 근사한 것도 그림에 나타나 있다(출처: Bayes, 1764, 385쪽).
    <베이즈의 당구대>
    (source: 조재근(역), 통계학의 역사 (한길사, 1986), p.278)

위의 실험을 보다 쉽게 설명하면 먼저 베이즈는 탁자를 등지고 서서 조수에게 첫 번째 공을 굴리게 한다. 탁자는 평평하기 때문에 그 공은 어느 지점에든 동일한 확률로 멈춘다. 하지만 그는 그 공이 어디에 멈춘지 알지 못한다. 그리고 두 번째 공을 굴리게 하고 멈춘 지점이 첫 번째 공보다 왼쪽에 있는지 오른쪽에 있는지만 알려 달라고 한다. 동일한 실험을 계속 반복하면 베이즈는 탁자를 등지고 서 있지만 첫번째 공이 멈춘 위치의 범위를 점점 더 좁혀나갈 수 있게 된다. 이렇게 두 번째 이후의 시도로서 첫 번째 공의 위치를 추적한다는 점에서 베이즈의 정리는 역확률, 원인의 확률이라고 불린다.

흄과 베이즈의 입장을 정리하면 다음과 같은 차이가 드러난다. 흄에게는 지금까지 매일 해가 떴다고 해도 내일 아침 해가 뜰 것이란 예측이 해가 떠 오르지 않을 것이란 예측보다 더 합리적이라고 보지 않는다. 그러나 베이즈 정리에서는 아침마다 해가 떠오르는 것을 봤기 때문에 매일 아침 해 뜨는 것이 영원한 특성일 것이라는 믿음이 점점 커간다. 그 특성에 100% 확신을 부여할 수는 없지만 그 지점으로 끊임없이 수렴한다. 신이 제1원인이라는 부분에 대해서도 베이즈 정리는 ‘결함투성이의 인간에게 오류가 있다고 해서 신에게 오류가 있다고 오해하지 말라’고 한다. 진리를 향해 끊임 없이 나아가면 반드시 신의 진실한 모습을 볼 수 있을 것이라고 믿는 것이다.

당구대 실험의 베이즈 정리는 몇 개의 구성 요소를 갖고 있다. 먼저 기존의 확률에 ‘사전(prior)’, 새로운 자료로 보강된 다른 가설의 확률에는 ‘우도(likelihood)’, 그리고 새로 수정한 확률에는 ‘사후(posterior)’라는 명칭을 부여한다. 새로운 시도를 할 때마다 새로운 추정이 진행되고 그 결과 얻어지는 사후 확률은 다음 시도에서 사전확률 역할을 한다. 이는 일종의 진화적 체계여서 새로운 시도가 있을 때마다 새로운 정보가 추가되어 확실성이 높아진다. 이를 요약하면 <‘이전’의 가능성은 ‘이후’에 비례한다>로 정리할 수 있다.

수학에 친하지 않은 사람들에게는 어렵게 보일 수 있지만 위의 내용을 수학적으로 정리한 가장 보편적인 베이즈 정리의 모습은 아래와 같다.

![베이지안수식](http://ncc.phinf.naver.net/20160808_246/1470620423744Hcy5I_JPEG/02.jpg)

P(A|B)는 조건부 확률이다. 즉 B라는 조건하에서 A가 일어날 확률을 의미한다. 위의 식에서 양변을 P(B|A)로 나누면 위에서 이야기한 이전의 확률이 이후의 확률에 비례한다는 공식을 얻을 수 있다.

베이즈 정리는 위의 공식에 나오는 P(A)라는 사전확률이 필요하다. 그리고 P(B)라는 우도를 계산해야 한다. 우도(likelihood probability)는 이미 사건 A가 일어난 상황에서 다른 사건들이 일어날 확률을 모두 더한 것이다. 이 사전확률과 우도가 지닌 학문적, 수학적 계산의 문제를 지난 250년동안 어떻게 해결해 왔는지를 보여 주는 것이 바로 저자 샤론 맥그레인(Sharon Bertsch McGrayne)이 <불멸의 이론>에서 다루는 내용이다. 두 문제의 해결책은 몬테카를로 시뮬레이션과 마르코프 사슬에 의해서 해결된다. 베이즈가 죽은 지 250년이 흐른 뒤 베이즈 이론의 취약점으로 지적되어 온 선험적 사전확률이라는 멍에와 변수가 증가하면서 수많은 사례들을 계산해야만 했던 우도 함수 문제에 대해 1989년 캐나다 퀘벡의 워크숍에서 스미스와 겔팬드(Smith & Gelfand)가 마르코프 사슬 몬테카를로(Markov chain Monte Carlo)을 통해 풀 수 있음을 보여주면서 베이지안 통계는 마침내 통계학의 패러다임을 바꾸게 된다.

##### 베이지안 통계의 활용

베이지안 통계가 오랫동안 학계에서 비학술적 통계 방법으로 취급받아 온 결정적 계기는 로널드 피셔(Ronald Fisher)가 고안한 빈도에 기반한 통계 방법론 때문이었다. 빈도주의 통계학은 지금도 사회과학계에서 널리 사용되고 있는 통계법이다. 무작위 표본추출, 유의성 검정, 최대우도측정법, 분산분석, 실험계획법 등이 그가 개척한 분야다. 정확한 실험실 데이터와 엄정한 수학적 도구를 활용하여 의문에 대한 해답을 찾아가는 그의 방법론으로 볼 때, 사전 확률과 같은 경험에 기반한 선험적 수치에 과학이라는 칭호를 부여하는 것은 용납할 수 없는 일이었다. 특정한 일이 발생할 확률을 동일하다고 가정한 (당구대 특정 위치에 공이 멈출 확률이 동일하다고 가정했던) 부분에 대해서도 피셔리안 통계학자들은 수긍할 수 없었다.

학계에서는 채택할 수 없는 방법론이라는 사망선고를 받았지만, 베이지안 통계는 정부와 산업계에서 불확실성에 마주칠 때마다 강력한 해법을 제시했다. 가령 냉전시대에 핵폭탄이 실수로 폭발할 가능성이 얼마인지 물었을 때, 피셔리안(Fisherian) 통계는 전혀 답을 내지 못했다. 핵폭탄이 잘못되어 폭발한 사례 자체가 많지 않았기 때문이다. 핵잠수함이 대서양 한 가운데서 침몰했을 때 그 위치를 찾고자 하는 정부에게 피셔리안 통계는 아무런 답을 내지 못했다. 미래 시장에서 어떤 신상품이 어느 정도의 성공확률이 있는지를 묻는 경영상의 의사결정에 대해서도 피셔리안 통계는 효율적인 답을 내 놓지 못했다. 산업화 초기 산재보험 요율을 계산할 때도 마찬가지로 선례가 많지 않았기 때문에 피셔리안 통계는 한계를 가지고 있었다.

2차 세계대전 때 독일의 에니그마 암호를 풀어내기 위해 튜링 팀이 동원한 방법이 바로 베이지안 통계였다. 암호화된 내용(결과)을 가지고 원래 내용(원인)을 찾아가는 일에 필요한 방법이 바로 베이지안 통계법이었기 때문이다.

베이지안 통계가 상대적으로 유용한 해법을 제시했던 분야의 공통점은 ‘불확실’성이 큰 영역이라는 점이었다. 불확실성을 공략하는 베이지안의 방법론은 ‘거친’ 추측과 상대적인 확률에 기반해 정보를 모으고 그 정보로 기존의 추측을 업데이트해 나가는 과정을 밟는다. 하지만 피셔리안 통계에서는 양질의 데이터가 충분하기 때문에 대부분의 불확실성이 상대적인 확률이 아니라 상대적인 ‘빈도’로 규정된다. 그래서 피셔리안 통계는 흔히 ‘빈도주의’ 통계라고도 불린다.

여기서 잠깐 <신호와 소음>에서 네이트 실버가 지적한 피셔리안 통계의 한계를 살펴보자. 그는 빈도주의 통계가 이론과 실제 모두에서 객관적이지 않다고 본다. 수많은 가정을 전제로 진행되기 때문이다. 먼저 빈도주의 방법론은 일반적으로 어떤 측정 속에 내재하는 불확실성은 종 모양의 정규분포를 따른다고 설정한다. 이런 설정은 어느 분야에서는 훌륭하게 들어 맞지만 주식시장의 변동은 설명하지 못한다. 둘째, 빈도주의 통계에서는 표본집단 설정을 전제로 하는데 여러 실용적 응용 분야에서는 완전히 임의적이다. 가령 9/11 테러 같은 경우 무엇을 표본집단으로 설정할 수 있는가? 셋째, 가장 재미있게도, 연구자들의 ‘편견’이 개입하는 것을 막는 것이 문제라고 본다. 즉 연구자가 자신이 설정한 가설에 내재하는 가능성 또는 전체 맥락 속의 의미를 고려하지 못하게 가로막는다는 것이다. 그 결과 두꺼비가 지진을 예측한다는 논문들이 통계적으로 유의미한 결과로 발표된다는 것이다. 네이트 실버는 여기서 더 나아가 현재 통계적 접근법의 패러다임이 바뀌고 있으며 피셔의 빈도주의 통계학을 학부에서 가르치지 말아야 한다고 주장하는 학자들이 많아지고 있고 가설검증을 학술지에서 추방할 것을 검토하는 움직임도 있다고 소개하며, 결국은 베이지안 통계가 빈도주의 통계를 누르고 승리할 것이라고 ‘예측’한다.

##### 총평

미루다 미루다 <불멸의 이론> 서평을 쓰게 된 동기는 먼저 우리 사회에도 베이지안 통계에 대한 세례가 필요하지 않을까 싶어서다. 베이지안 방법론은 250년전 영국에서 태어나 프랑스의 라플라스를 거쳐 이론적 모습을 갖췄다. 베이지안은 2차 세계대전 때 다시 영국에서 크게 활용되지만 전쟁자료의 비밀문서화 정책에 따라 다시 묻힌다. 그러나 미국은 베이지안 방법론을 다양하게 활용하며 많은 사회적 문제와 경영의 의사결정, 정책 문제 등을 풀어 나간다. 그래서 베이지안 방법론에서는 미국의 실용주의 냄새가 많이 느껴진다. 마르코프 체인 몬테카를로 방법론이 나오기 전에도 여러 가지 제한점이 있음에도 불구하고 베이지안 방법론이 다양한 영역에서 채택되었다.

이 서평을 쓰게 된 두 번째 이유는 베이지안 방법론이 인생을 살아가는 좋은 인생관을 제공해 준다는 점이다. 베이지안 방법론은 실패했다고 주저앉지 않고 끊임 없이 다음 번 시도를 행할 이유를 제공해 준다. 매번의 새로운 시도에서 배우는 지식이 삶의 지혜로 나아가는 과정을 보여주기 때문이다. 지상에서 벌어지는 모든 오류와 부조리가 신의 부재를 증명하는 것이 아니라 신을 찾아가는 여정에서 만나는 정보의 일종이며 끊임 없이 신을 향해 나아가다 보면 점점 더 진실에 다가가게 된다는 베이즈 정리의 이론적 구조는 베이즈가 흄의 회의론에 가한 통쾌한 일격이다.

마지막 이유는 ICT 영역에서 화두가 된 알고리즘의 진화와 기계학습, 인공지능 등이 모두 베이지안 방법론에 근간을 두고 있다는 점이다. 데이터가 쌓이면서 타겟팅하는 고객이 더욱 정교해 지고, 음성 인식률이 올라가며, 추천은 더욱 더 맞춤화 되어 간다. 기계의 지능이 인간을 앞지르게 되는 특이점(Singularity)도 베이지안 방법론에 기반한 인공지능의 발전 때문에 생길 것이다.

저자 샤론 맥그레인은 <불멸의 이론>을 학술적 내용으로 기술하지는 않았다. 대신 책의 주요 내용은 베이지안 방법론이 어떤 맥락에서 어떻게 적용되었는지를 중심으로 기술한다. (역확률 혹은 원인의 확률에 대한 수학적 설명을 보고 싶으면 <통계학의 역사>를 참조하시길 권한다.) 이 부분이 이 책의 강점과 약점을 동시에 보여주는 지점이다. 어떤 수학적 계산을 거쳐 어떻게 결과를 도출했는지에 대한 자세한 설명이 없기 때문에 일반 독자들이 책을 쉽게 집어들 수 있는 만큼 조금 더 깊고 실용적인 수학적 문제 해결법을 찾는 사람들에게 이 책은 약간의 갈증을 남겨 줄 책이다. 그럼에도 불구하고 하나의 이론이 만들어진 후 약 250년의 세월 동안 일어났던 일을 한 권의 책으로 읽을 수 있고 인생과 산업계를 조망할 기회를 준다는 점에서 일독을 권하는 바이다.
